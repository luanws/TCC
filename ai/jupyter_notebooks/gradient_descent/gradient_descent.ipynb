```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_directory(path):
    if not os.path.exists(path):
        os.makedirs(path)
create_directory('jupyter_notebooks/gradient_descent/img')
```

Descida do gradiente

```python
def f(x):
    return x**2


def diff_f(x):
    return 2*x


def draw_point(x, *args, **kwargs):
    plt.plot(x, f(x), 'ro', *args, **kwargs)


def draw_gradient_descent_arrow(x, *args, **kwargs):
    plt.quiver(x, f(x), -1, -diff_f(x), angles='xy', scale_units='xy', scale=1, *args, **kwargs)


point = 2
x = np.linspace(-3, 3, 1000)
plt.plot(x, f(x), 'g')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
draw_point(point)
draw_gradient_descent_arrow(point)
plt.savefig('jupyter_notebooks/gradient_descent/img/Descida do gradiente.jpg', dpi=300)
plt.show()
```

Descida do gradiente com mínimo local

```python
def f(x):
    return x**2 + 10*np.sin(x)


def diff_f(x):
    return 2*x + 10*np.cos(x)


point = 4
x = np.linspace(-4, 6, 1000)
plt.plot(x, f(x), 'g')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
draw_point(point)
plt.savefig('jupyter_notebooks/gradient_descent/img/Descida do gradiente com mínimo local.jpg', dpi=300)
plt.show()
```
